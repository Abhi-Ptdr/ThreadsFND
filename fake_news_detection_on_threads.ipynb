# **Data Preprocessing**


## Text Normalization and Text Cleaning

```python
# Convert all text to lowercase:

df['Post Text'] = df['Post Text'].str.lower()
```

```python
# Remove special characters, URLs, and unnecessary whitespace:
import re
df['Post Text'] = df['Post Text'].apply(lambda x: re.sub(r'http\S+|www.\S+', '', x))  # Remove URLs
df['Post Text'] = df['Post Text'].apply(lambda x: re.sub(r'[^\w\s]', '', x))         # Remove special chars
df['Post Text'] = df['Post Text'].str.strip()                                        # Remove extra spaces
```

```python
# Normalize Unicode Characters: Some characters might appear as special symbols (e.g., Â, Ã) due to encoding issues. Normalize your text to remove these artifacts:

import unicodedata

def normalize_text(text):
    # Normalize Unicode characters to ASCII and remove non-ASCII
    return unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')

df['Post Text'] = df['Post Text'].apply(normalize_text)
```

```python
# Remove Non-Alphabetic Characters: After normalization, you can further clean the text by keeping only alphabetic characters and spaces:

df['Post Text'] = df['Post Text'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
```

```python
# Check for Empty Strings: After cleaning, some rows may have empty or meaningless text. Remove them

df = df[df['Post Text'].str.strip() != '']
```

```python
# Remove Repeated Characters: Words like "aaa", "aa", and "aaaa" often result from typos or noise.

import re

def remove_repeated_characters(text):
    # Replace sequences of the same character repeated 2+ times (e.g., "aaa") with a single character
    return re.sub(r'\b(\w)\1{2,}\b', '', text)  # Matches "aaa", "eeee", etc.

df['Post Text'] = df['Post Text'].apply(remove_repeated_characters)
```

```python
# Remove Short Words: Words that are too short (e.g., "aa", "a") often don't carry meaning.

df['Post Text'] = df['Post Text'].apply(lambda x: ' '.join([word for word in x.split() if len(word) > 2]))
```

### Sentiment analysis of fake posts:
Are they generally more negative or sensationalized?

The compound score is a normalized score calculated by combining the individual positive, neutral, and negative scores.
It represents the overall sentiment polarity of the text.
The value ranges from -1 (very negative) to 1 (very positive).

Compound Score	Interpretation
≤ -0.05	Negative Sentiment 😡
-0.05 to 0.05	Neutral Sentiment 😐
≥ 0.05	Positive Sentiment 😊

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Download VADER lexicon if not already downloaded
nltk.download("vader_lexicon")

# Initialize Sentiment Analyzer
sia = SentimentIntensityAnalyzer()

# Function to analyze sentiment
def get_sentiment_scores(text):
    if pd.isna(text):  # Handle missing values
        return pd.Series([0, 0, 0, 0])
    scores = sia.polarity_scores(text)
    return pd.Series([scores["neg"], scores["neu"], scores["pos"], scores["compound"]])

# Apply sentiment analysis to Post Text
df[["Negative", "Neutral", "Positive", "Compound"]] = df["Post Text"].apply(get_sentiment_scores)

# Separate fake and real posts
fake_posts = df[df["label"] == 1]
real_posts = df[df["label"] == 0]

# Compare Average Sentiment Scores ---
avg_sentiment = df.groupby("label")[["Negative", "Neutral", "Positive", "Compound"]].mean()
avg_sentiment.index = ["Real Posts", "Fake Posts"]

# Print the sentiment scores
print("🔹 Average Sentiment Scores for Fake vs. Real Posts:")
print(avg_sentiment)

# Plot Sentiment Scores
plt.figure(figsize=(10, 5))
ax = avg_sentiment.plot(kind="bar", colormap="coolwarm", width=0.7)

# Annotate bars with sentiment scores
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", label_type="edge", fontsize=12, padding=3)

# Customize Plot
plt.title("Average Sentiment Scores: Fake vs. Real Posts")
plt.xlabel("Post Type")
plt.ylabel("Sentiment Score")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show Plot
plt.show()
```

### Analyze bot-like behavior and network interactions of fake news spreaders

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# Load dataset (ensure 'Username', 'Post Date', 'Post Text', 'label', 'Likes', 'Replies', 'Reposts', 'Shares' exist)
# df = pd.read_csv("your_data.csv")

### 1. Detecting Bot-like Behavior ###

# Convert 'Post Date' to datetime format
# df["Post Date"] = pd.to_datetime(df["Post Date"])

# Count posts per user
user_post_counts = df.groupby("Username")["Post Date"].count().reset_index()
user_post_counts.columns = ["Username", "Total Posts"]

# Identify potential bots (users posting excessively)
threshold = user_post_counts["Total Posts"].quantile(0.95)  # Top 5% most frequent posters
potential_bots = user_post_counts[user_post_counts["Total Posts"] > threshold]
print(f"Potential bot accounts (posting too frequently):\n{potential_bots}")

# Identify users posting at unusual times (e.g., 24/7 activity)
df["Post Hour"] = df["Post Date"].dt.hour
activity_by_hour = df.groupby(["Username", "Post Hour"]).size().unstack(fill_value=0)
active_hours = (activity_by_hour > 0).sum(axis=1)
bots_posting_all_hours = active_hours[active_hours >= 18]  # Users posting at least 18 hours a day
print(f"Users posting 24/7 (possible bots):\n{bots_posting_all_hours}")

### 2. Network Analysis of Fake News Spreaders ###

# Filter dataset for fake news (label = 1)
fake_news_df = df[df["label"] == 1]

# Create interaction network: Nodes (Users), Edges (Interactions)
G = nx.Graph()

# Add nodes (Users)
users = fake_news_df["Username"].unique()
G.add_nodes_from(users)

# Add edges based on interactions (likes, replies, reposts, shares)
for _, row in fake_news_df.iterrows():
    user = row["Username"]

    # Example: Add weighted edges based on engagement
    G.add_edge(user, f"Likes_{row['Likes']}", weight=row["Likes"])
    G.add_edge(user, f"Replies_{row['Replies']}", weight=row["Replies"])
    G.add_edge(user, f"Reposts_{row['Reposts']}", weight=row["Reposts"])
    G.add_edge(user, f"Shares_{row['Shares']}", weight=row["Shares"])

# Identify influential users using PageRank
pagerank_scores = nx.pagerank(G, alpha=0.85)
top_users = sorted(pagerank_scores.items(), key=lambda x: x[1], reverse=True)[:10]
print(f"Top 10 most influential fake news spreaders:\n{top_users}")

# Visualize the interaction network
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)  # Node positions
nx.draw_networkx_nodes(G, pos, node_size=50, alpha=0.7)
nx.draw_networkx_edges(G, pos, alpha=0.3)
nx.draw_networkx_labels(G, pos, font_size=7)
plt.title("Fake News Spreaders Network")
plt.show()
```

#### Enhanced Bot Detection

```python
import pandas as pd
import numpy as np
import networkx as nx
from collections import defaultdict, Counter
from difflib import SequenceMatcher
import matplotlib.pyplot as plt
import seaborn as sns

#########################
# 1. Repetitive Content #
#########################
def detect_repetitive_content(df, text_col="Post Text", username_col="Username"):
    """
    Flags users who have multiple (e.g., 3) pairs of consecutive posts
    with >90% similarity.
    """
    similarity_threshold = 0.9
    min_repetitions = 3  # Require at least 3 repeated pairs

    user_texts = defaultdict(list)
    for _, row in df.iterrows():
        user_texts[row[username_col]].append(row[text_col])

    repetitive_users = set()

    for user, posts in user_texts.items():
        repeated_pairs = 0
        for i in range(len(posts) - 1):
            similarity = SequenceMatcher(None, posts[i], posts[i + 1]).ratio()
            if similarity > similarity_threshold:
                repeated_pairs += 1
        if repeated_pairs >= min_repetitions:
            repetitive_users.add(user)

    return repetitive_users

########################
# 2. Burst Posting     #
########################
def detect_burst_posting(df, time_col="Post Time (HH:MM)", username_col="Username"):
    """
    Flags users who have more than 10 burst posts
    (posts within 10 seconds of each other).
    """
    df = df.copy()
    df[time_col] = pd.to_datetime(df[time_col], errors="coerce")  # Convert to datetime
    # Filter out any NaT rows (invalid time format)
    df = df.dropna(subset=[time_col])

    df = df.sort_values(by=[username_col, time_col])  # Sort by user and time

    burst_users = set()
    burst_threshold = pd.Timedelta(seconds=10)
    min_bursts = 10  # More than 10 burst posts

    for user, group in df.groupby(username_col):
        time_diffs = group[time_col].diff().dropna()
        # Count how many times the diff is <= 10 seconds
        burst_count = (time_diffs <= burst_threshold).sum()
        if burst_count > min_bursts:
            burst_users.add(user)

    return burst_users

###########################################
# 3. High Centrality in Engagement Graph  #
###########################################
def detect_high_centrality_users(df, username_col="Username"):
    """
    Flags top 10 users with highest degree centrality in
    the engagement network (Likes, Replies, Reposts, Shares).
    """
    G = nx.Graph()

    for _, row in df.iterrows():
        for engagement_type in ["Likes", "Replies", "Reposts", "Shares"]:
            # Convert to numeric if needed
            value = 0
            try:
                value = float(row[engagement_type])
            except:
                continue

            if value > 0:  # Only add edges if there's engagement
                G.add_edge(row[username_col], f"{engagement_type}_{value}")

    if len(G.nodes()) == 0:
        return set()

    centrality_scores = nx.degree_centrality(G)
    sorted_centrality = sorted(centrality_scores.items(), key=lambda x: x[1], reverse=True)

    # Extract top 10 users (if they are actual usernames, not "Likes_XXX" nodes)
    top_spreaders = []
    for user, score in sorted_centrality:
        if user in df[username_col].values:
            top_spreaders.append(user)
        if len(top_spreaders) == 10:
            break

    return set(top_spreaders)

###############################
# 4. Combine Criteria Strictly#
###############################
def label_bots_strict(df):
    """
    Labels a user as bot if they meet >=2 criteria.
    """
    repetitive_users = detect_repetitive_content(df)
    burst_users = detect_burst_posting(df)
    high_centrality_users = detect_high_centrality_users(df)

    # Create a dict to count how many criteria each user meets
    criteria_count = Counter()

    # For each user flagged by each criterion, add 1
    for user in repetitive_users:
        criteria_count[user] += 1
    for user in burst_users:
        criteria_count[user] += 1
    for user in high_centrality_users:
        criteria_count[user] += 1

    # Label users as bots if they meet at least 2 criteria
    potential_bots_strict = {user for user, count in criteria_count.items() if count >= 2}

    return potential_bots_strict, (repetitive_users, burst_users, high_centrality_users)

##################################
#        MAIN ANALYSIS           #
##################################
if __name__ == "__main__":
    # df = pd.read_csv("your_fake_news_dataset.csv")

    potential_bots_strict, all_sets = label_bots_strict(df)
    repetitive_users, burst_users, high_centrality_users = all_sets

    print(f"Users flagged for Repetitive Content: {len(repetitive_users)}")
    print(f"Users flagged for Burst Posting: {len(burst_users)}")
    print(f"Users flagged for High Centrality: {len(high_centrality_users)}")
    print("---------------------------------------------------")
    print(f"Total Bots (meeting >= 2 criteria): {len(potential_bots_strict)}")
    print(f"Example Bot-Like Users (strict): {list(potential_bots_strict)[:10]}")

    # Create Is_Bot column for strict labeling
    df["Is_Bot"] = df["Username"].apply(lambda x: x in potential_bots_strict)
    df.to_csv("refined_bot_detection_strict.csv", index=False)

    # Visualization: Post Frequency of Bots vs. Humans
    # Convert Post Date to datetime if needed
    df["Post Date"] = pd.to_datetime(df["Post Date"], errors="coerce")
    df = df.dropna(subset=["Post Date"])

    plt.figure(figsize=(10, 5))
    sns.histplot(df[df["Is_Bot"]]["Post Date"], bins=30, color="red", label="Bots", kde=True)
    sns.histplot(df[~df["Is_Bot"]]["Post Date"], bins=30, color="blue", label="Humans", kde=True)
    plt.legend()
    plt.xlabel("Post Date")
    plt.ylabel("Frequency")
    plt.title("Posting Frequency: Bots vs. Humans (Strict Criteria)")
    plt.xticks(rotation=45)
    plt.show()
```

### Naive Bayes

```python
import time
from sklearn.naive_bayes import ComplementNB
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Scale SBERT embeddings to be non-negative
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define parameter grid
param_grid = {
    'alpha': np.logspace(-3, 1, 10),
    'norm': [True, False]
}

cnb_model = ComplementNB()
stratified_kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

# --- Measure time for hyperparameter tuning ---
start_time = time.time()

cnb_random = RandomizedSearchCV(
    estimator=cnb_model,
    param_distributions=param_grid,
    n_iter=20,
    cv=stratified_kfold,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
cnb_random.fit(X_train_scaled, y_train)

end_time = time.time()
print(f"\nTime taken for hyperparameter tuning: {end_time - start_time:.2f} seconds")

# Best hyperparameters
best_cnb_params = cnb_random.best_params_
print("Best Hyperparameters for ComplementNB:", best_cnb_params)

# --- Measure time for training final model ---
start_train_time = time.time()
best_cnb = ComplementNB(**best_cnb_params)
best_cnb.fit(X_train_scaled, y_train)
end_train_time = time.time()
print(f"Time taken to train best ComplementNB model: {end_train_time - start_train_time:.2f} seconds")

# Predict and evaluate
y_pred_cnb = best_cnb.predict(X_test_scaled)
accuracy_cnb = accuracy_score(y_test, y_pred_cnb)
precision_cnb = precision_score(y_test, y_pred_cnb, average='weighted')
recall_cnb = recall_score(y_test, y_pred_cnb, average='weighted')
f1_cnb = f1_score(y_test, y_pred_cnb, average='weighted')

print(f"\nComplementNB Accuracy: {accuracy_cnb:.4f}")
print(f"ComplementNB Precision: {precision_cnb:.4f}")
print(f"ComplementNB Recall: {recall_cnb:.4f}")
print(f"ComplementNB F1-score: {f1_cnb:.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred_cnb))
```

### AdaBoost
Combines Synthetic Minority Over-sampling Technique (SMOTE) with AdaBoost. In each iteration:

SMOTE generates synthetic samples for the minority class.

```python
import pandas as pd
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration
from IPython.display import FileLink

# Load the dataset
dataset_path = "/kaggle/working/Cleaned_NewThreadsData_Labeled.csv"
df = pd.read_csv(dataset_path, encoding='ISO-8859-1')

# Ensure required columns exist
assert "Post Text" in df.columns and "label" in df.columns, "Dataset must have 'Post Text' and 'label' columns."

# Filter only fake news samples
fake_news_df = df[df['label'] == 1]["Post Text"].dropna().tolist()

# Load T5-Large model and tokenizer
model_name = "t5-large"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name).to("cuda" if torch.cuda.is_available() else "cpu")

def generate_manipulated_fake_news(texts, num_variations=3):
    manipulated_samples = []
    for text in texts:
        # Improved prompt to encourage slight modification while preserving meaning
        prompt = f"Paraphrase the following news post while keeping its meaning intact. Make slight word and phrasing changes: {text}"
        inputs = tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True).to(model.device)

        # Generate 3 variations per original text
        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_length=512,
                num_return_sequences=num_variations,  # Generate 3 versions
                temperature=0.7,   # Moderate randomness
                top_k=50,          # Top 50 token choices for better variation
                top_p=0.95,        # Nucleus sampling to avoid extreme changes
                do_sample=True
            )

        # Decode and store multiple variations
        for output in outputs:
            manipulated_text = tokenizer.decode(output, skip_special_tokens=True)
            manipulated_samples.append((text, manipulated_text))  # Store both original and manipulated

    return manipulated_samples

# Generate manipulated fake news (with 3 variations per original)
num_samples_needed = 1500
num_samples_to_generate = 1500  # Ensure we don't exceed available samples
selected_fake_news = fake_news_df[:num_samples_to_generate]
manipulated_fake_news_pairs = generate_manipulated_fake_news(selected_fake_news, num_variations=3)

# Print some original vs manipulated fake news examples
print("\nSample Manipulated Fake News:")
for i in range(10):  # Print first 10 sets of variations
    original = manipulated_fake_news_pairs[i * 3][0]  # Original text
    variations = [manipulated_fake_news_pairs[i * 3 + j][1] for j in range(3)]  # Get 3 variations
    print(f"Original: {original}")
    for idx, variant in enumerate(variations, 1):
        print(f"Variation {idx}: {variant}")
    print("-" * 80)

# Create a new DataFrame with all manipulated fake news
new_fake_news_df = pd.DataFrame({"Post Text": [pair[1] for pair in manipulated_fake_news_pairs], "label": 1})

# Merge with the original dataset
df = pd.concat([df, new_fake_news_df], ignore_index=True)

# Shuffle dataset
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Save the balanced dataset
balanced_dataset_path = "balanced_dataset_with_T5Large.csv"
df.to_csv(balanced_dataset_path, index=False)
print(f"\nBalanced dataset saved at {balanced_dataset_path}")

# Provide download link in Kaggle notebook
print("\nClick below to download the dataset:")
display(FileLink(balanced_dataset_path))
```

```python
# Convert "Post Text" column to string data type
df['Post Text'] = df['Post Text'].astype(str)

# Verify the conversion
print(df['Post Text'].dtype)


# Define the conversion function again (if not already applied)
def convert_k_to_numeric(value):
    if value is None or pd.isna(value):  # Handle None or NaN
        return np.nan
    if isinstance(value, str) and 'k' in value.lower():  # Handle "k" (case insensitive)
        return float(value.replace('k', '').replace('K', '')) * 1000
    try:
        return float(value)  # Convert other values to float
    except ValueError:
        return np.nan

# Apply the conversion to all relevant columns
columns_to_convert = ['Likes', 'Replies', 'Reposts', 'Shares']
for column in columns_to_convert:
    df[column] = df[column].apply(convert_k_to_numeric)

# Check if columns are properly converted
print(df[columns_to_convert].dtypes)


# Convert Post Date and Post Time to datetime for analysis
# df['Post Date'] = pd.to_datetime(df['Post Date'])
# df['Post Time'] = pd.to_datetime(df['Post Time (HH:MM)'], format='%H:%M').dt.time

# # Extract hour from Post Time
# df['Post Hour'] = df['Post Time'].apply(lambda x: x.hour)
```

```python
# Convert all text to lowercase:

df['Post Text'] = df['Post Text'].str.lower()
```

```python
# Remove special characters, URLs, and unnecessary whitespace:
import re
df['Post Text'] = df['Post Text'].apply(lambda x: re.sub(r'http\S+|www.\S+', '', x))  # Remove URLs
df['Post Text'] = df['Post Text'].apply(lambda x: re.sub(r'[^\w\s]', '', x))         # Remove special chars
df['Post Text'] = df['Post Text'].str.strip()                                        # Remove extra spaces
```

```python
# Normalize Unicode Characters: Some characters might appear as special symbols (e.g., Â, Ã) due to encoding issues. Normalize your text to remove these artifacts:

import unicodedata

def normalize_text(text):
    # Normalize Unicode characters to ASCII and remove non-ASCII
    return unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')

df['Post Text'] = df['Post Text'].apply(normalize_text)
```

```python
# Remove Non-Alphabetic Characters: After normalization, you can further clean the text by keeping only alphabetic characters and spaces:

df['Post Text'] = df['Post Text'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
```

```python
# Check for Empty Strings: After cleaning, some rows may have empty or meaningless text. Remove them

df = df[df['Post Text'].str.strip() != '']
```

```python
# Remove Repeated Characters: Words like "aaa", "aa", and "aaaa" often result from typos or noise.

import re

def remove_repeated_characters(text):
    # Replace sequences of the same character repeated 2+ times (e.g., "aaa") with a single character
    return re.sub(r'\b(\w)\1{2,}\b', '', text)  # Matches "aaa", "eeee", etc.

df['Post Text'] = df['Post Text'].apply(remove_repeated_characters)
```

```python
# Remove Short Words: Words that are too short (e.g., "aa", "a") often don't carry meaning.

df['Post Text'] = df['Post Text'].apply(lambda x: ' '.join([word for word in x.split() if len(word) > 2]))
```

### AdaBoost
Combines Synthetic Minority Over-sampling Technique (SMOTE) with AdaBoost. In each iteration:

SMOTE generates synthetic samples for the minority class.
The AdaBoost algorithm is applied to the newly balanced dataset.

```python
import time
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Define parameter grid for AdaBoost
param_grid = {
    'n_estimators': [50, 100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.5],
    'algorithm': ['SAMME', 'SAMME.R'],
}

# Initialize AdaBoost Classifier
adaboost = AdaBoostClassifier(random_state=42)

# Stratified K-Fold
stratified_kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

# Randomized Search
adaboost_random = RandomizedSearchCV(
    estimator=adaboost,
    param_distributions=param_grid,
    n_iter=20,
    cv=stratified_kfold,
    verbose=2,
    random_state=42,
    n_jobs=-1
)

# Apply SMOTE
smote = SMOTE(sampling_strategy=1.0, k_neighbors=5, random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Fit model and record training time
start_time = time.time()
adaboost_random.fit(X_train_resampled, y_train_resampled)
end_time = time.time()

print(f"\n Hyperparameter Search Time: {end_time - start_time:.2f} seconds")

# Output best parameters
best_adaboost_params = adaboost_random.best_params_
print("\nBest Hyperparameters for AdaBoost:", best_adaboost_params)
```

```python
import time
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Train the best model with resampled data
best_adaboost = AdaBoostClassifier(**best_adaboost_params, random_state=42)

start_time = time.time()
best_adaboost.fit(X_train_resampled, y_train_resampled)
end_time = time.time()

print(f"\n Training Time: {end_time - start_time:.2f} seconds")

# Predict on test set
y_pred_adaboost = best_adaboost.predict(X_test)

# Evaluate Performance
accuracy_adaboost = accuracy_score(y_test, y_pred_adaboost)
precision_adaboost = precision_score(y_test, y_pred_adaboost, average='weighted')
recall_adaboost = recall_score(y_test, y_pred_adaboost, average='weighted')
f1_adaboost = f1_score(y_test, y_pred_adaboost, average='weighted')

# Print results
print(f"\nAdaBoost Accuracy: {accuracy_adaboost:.4f}")
print(f"AdaBoost Precision: {precision_adaboost:.4f}")
print(f"AdaBoost Recall: {recall_adaboost:.4f}")
print(f"AdaBoost F1-score: {f1_adaboost:.4f}")

# Detailed classification report
print("\nAdaBoost Classification Report:\n", classification_report(y_test, y_pred_adaboost))
```
